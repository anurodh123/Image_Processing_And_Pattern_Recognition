```
# Tuples (inmutable,ordered)
# Create a tuple of days in a week

# Using +ve logic

days = ("Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday")
print("weekend : ",days[5:7])
print("weekdays : ",days[:5])
print("First day : ",days[0])

# Using -ve logic

print("Last day : ",days[-1])
print("weekdays : ",days[::-1])
print("weekend : ",days[-2:])
print("weekdays : ",days[:-2])
```
